##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np 
arr = np.array([1,3,4,5,9])
print(arr[0])
print(arr[[0,2,3]])
```

You can do this fancy indexing in 2D Matrix aswell

```python
arr = np.array([[1,2,4],[5,6,7]])
print(arr[[1,0],[1,2]])
print(arr[0,1])
```

##Conditional Selection
Find values based on conditions or replace them.

```python
arr = np.array([12,13,80,56,78,45,89,33,45,67,26])

np.where(arr>15,arr,6)

```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
arr = np.array([5,3,4,9,7,2,1,85,34,56,78])
np.sort(arr)
np.argsort(arr)
```

##Concatenation & Stacking

```python
arr1 = np.array([2,3,4])
arr2 = np.array([5,6,7])

np.concat([arr1,arr2])
```
